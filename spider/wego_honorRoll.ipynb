```python
from time import sleep
from bs4 import BeautifulSoup
from selenium import webdriver

option = webdriver.ChromeOptions() # 設定 chromedriver
# option.add_argument('--headless') # 無頭模式，開發完成之後再使用，可以完全背景執行，有機會變快。但無頭模式在某些網站會不能爬。 開發的時候先不要用
option.add_experimental_option('excludeSwitches', ['enable-automation']) # 開發者模式。可以避開某些防爬機制，有開有保佑 (本名不是開發者模式)
driver = webdriver.Chrome(options=option) # 啟動 chromedriver
driver.maximize_window() # 視窗放到最大。有些網站的版面會隨視窗大小改變，在沒有放到最大的情況下有可能定位正確但戳到其他地方，所以通常會放大
```

```python
url = 'https://www.wghs.tp.edu.tw/nss/p/index' # 指定網址
driver.get(url) # 進入指定網址
driver.implicitly_wait(5) # 等待伺服器反應最多 5 秒，如果在時間到之前反應就提早結束等待
```

```python
# 以現在台鐵時刻表的例子，出發站使用第一種語法如下：
sleep(1)
# 出發站的列表圖示
print(driver.page_source)
driver.find_element('css selector', '#middle > div > div > div:nth-child(1) > div > div > div:nth-child(2) > div > div.ct-row.gd-xs12 > div > ul > li:nth-child(2) > a').click()
sleep(1)
driver.forward() 
driver.quit()
```
