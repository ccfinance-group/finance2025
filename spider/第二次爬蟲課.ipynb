## 網頁編碼
每個網頁都會設定編碼方式，requests 預設網頁的編碼方式是 UTF8，中文網站比較可能是 UTF8 或是 BIG5，有時候抓到的網頁原始碼是亂碼，就是因為編碼方式不對。例如抓台灣證券交易所的最新訊息 https://www.twse.com.tw/zh/products/securities/etf/news.html

````python
from os import listdir
import requests
from bs4 import BeautifulSoup

chrome_version = listdir(r'C:\Program Files (x86)\Google\Chrome\Application')[0].split('.')[0]
header = {'User-Agent': f'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/{chrome_version}.0.0.0 Safari/537.36'}

url = 'https://www.twse.com.tw/zh/products/securities/etf/news.html'
r = requests.get(url, headers=header)
soup = BeautifulSoup(r.text, 'html.parser')
soup
````

````python
# 會發現標籤部分正常，但內容部分是亂碼，雖然可以看到在<head>中<meta charset="utf-8"/>表示網頁設定的編碼方式是 UTF8，
# 但基於某些神秘的原因解讀出來是亂碼。這時候只要再指定一次網頁的編碼方式是 UTF8 就可以解決了。

from os import listdir
import requests
from bs4 import BeautifulSoup

chrome_version = listdir(r'C:\Program Files (x86)\Google\Chrome\Application')[0].split('.')[0]
header = {'User-Agent': f'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/{chrome_version}.0.0.0 Safari/537.36'}

url = 'https://www.twse.com.tw/zh/products/securities/etf/news.html'
r = requests.get(url, headers=header)
r.encoding = 'UTF-8' # 指定編碼方式，utf-8、UTF-8、UTF8、utf8 都一樣
soup = BeautifulSoup(r.text, 'html.parser')
soup

# 但沒有抓到資料
````

## POST 請求
在 requests 爬蟲的 SOP 裡面，送請求的語法是 r = requests.get(url)，當我們不需要給網頁資料、直接跟網頁拿資料的時候，使用的是 get；當需要送一些資料，讓網頁伺服器決定要從資料庫撈什麼資料回傳的時候，使用的是 post。

例如要在台鐵時刻表( https://www.railway.gov.tw/tra-tip-web/tip/tip001/tip112/gobytime )取得今天從台北車站到台中車站的班次。

先在出發站和抵達站選好，然後打開開發人員工具的 Network，再按查詢，第一個跳出的就是你對網頁送的 POST 請求的資料。先看標頭(header)，可以在要求方法看到這是一個 POST 請求，而網址也跟原本的網址不一樣。把這個網址複製下來貼到 url，然後去看酬載(payload)，把裡面的每一個變數名稱和數值抄到程式碼裡，並整理成字典，結果如下面的程式碼：


````python
url = 'https://www.railway.gov.tw/tra-tip-web/tip/tip001/tip112/querybytime'
data = {'startStation': '1000-臺北', 'endStation': '3300-臺中', 'transfer': 'ONE', 'rideDate': '2025/05/28',
        'startOrEndTime': 'true', 'startTime': '00:00', 'endTime': '23:59', 'trainTypeList': 'ALL',
        'queryClassification': 'NORMAL', 'query': '查詢'}
r = requests.post(url, headers=header, data=data)
soup = BeautifulSoup(r.text, 'html.parser')
soup
````

````python
# 確實有想要的資料
soup.select('tbody tr[class="trip-column"] td:nth-child(2)')
````

可以觀察到，使用 post 的時候，多了一個 data 參數要填，這個 data 的內容就是我們剛剛抄下來的那些變數。其他部分都跟 get 請求一樣。現在可以看到 soup 內容就包含我們要的車次資料。只不過需要再想辦法定位出我們要的資料。

我們在用 python 寫函式的時候，會把一些變數設定預設值，在沒有資料輸入函式的時候，會自動填入這些預設值，網站伺服器也一樣。因此 POST 請求 data 裡面的變數可以只填一部分，但是哪些可以省略、哪些是必要的，必須要逐個刪掉測試才會知道。以台鐵時刻表來說，最精簡的變數如下：

### css selector 2
語法規則很多種，以爬蟲來說，爬比較新的網站，只要會 3 種就可以打天下。
1. 標籤  p # 抓標籤下的所有層的 p 標籤。
2. 標籤 > p # 抓標籤下一層的所有 p 標籤。
3. 標籤 p:nth-child(n) # 抓標籤下的第 n 個 p 標籤。從 1 開始

#### 標籤可以做限定
比不指定更能縮限範圍，在此只是以class這種屬性舉例，可以指定其他屬性。
1. 標籤[class="字串"] # 抓 class 為指定字串的所有標籤
2. 標籤[class*="字串"] # 抓 class 包含指定字串的所有標籤
3. 標籤[class^="字串"] # 抓 class 以指定字串開頭的所有標籤
4. 標籤[class$="字串"] # 抓 class 以指定字串結尾的所有標籤

組合起來會像是：標籤[class="字串"] p

````python
# 利用第一種把剛剛抓到的資料定位出來

data = {}
for row in soup.select('tbody tr[class="trip-column"]'):
    #因為要取全部所以取'td li'就好了 用text取純文字 然後把換行符號刪掉(用replace)
    train_number = row.select_one('td li').text.replace('\n', '')
    start_time = row.select('td')[1].text
    end_time = row.select('td')[2].text
    duration = row.select('td')[3].text
    data[train_number] = [train_number, start_time, end_time, duration]
    # break

data
````

````python
# 利用第三種把剛剛抓到的資料定位出來

data = {}
for row in soup.select('tbody tr[class="trip-column"]'):
    #因為要取全部所以取'td li'就好了 用text取純文字 然後把換行符號刪掉(用replace)
    #select_one比select快
    train_number = row.select_one('td li').text.replace('\n', '')
    #nth-child裡面的數字是從1開始
    start_time = row.select_one('td:nth-child(2)').text
    end_time = row.select_one('td:nth-child(3)').text
    duration = row.select_one('td:nth-child(4)').text
    data[train_number] = [train_number, start_time, end_time, duration]
    # break

data
````

## 使用API
API 的中文是應用程式介面，是網站的資料庫提供資料的一種管道。可以是網站自用，也可以提供給外面的人用。
+ 對網站來說，不會有爬蟲去佔網頁的頻寬；對爬蟲來說，有更方便取得資料的管道。雙贏。
+ 有些 API 需要申請且綁定帳號，這類的 API 對於網站來說是一種可以賣錢的服務，例如 OpenAI 和 google map。有些網站在設計上，前端會打 API 跟資料庫拿資料，我們可以在開發人員工具的 Network 裡面找到這樣的 API 網址，直接用這個網址跟資料庫拿資料，而不需要等網頁前端把東西都擺好之後才抓，速度會快很多。
+ 通常利用 API 爬蟲要用 requests 送請求，伺服器通常回傳 json 格式，所以要再用 json 轉成字典。

## 3 個爬蟲模組比較
在爬蟲專案常用模組的時候，提到了 pandas、requests、selenium，這三個模組各有適合的使用情境。
### pandas
+ 主要用於資料分析，爬蟲功能僅限於抓表格
  1. 直接送檔案的用 read_csv，政府的開放資料大多可以這樣抓
  2. 表格在網頁裡的用 read_html，可以直接這樣抓的網頁越來越少了，但可搭配 requests 或 selenium 使用
+ 可以用的話，爬回來的資料就是可以用 pandas 繼續處理的格式，很方便。

### requests
+ 模擬瀏覽器的行為，主要用 get 和 post 抓資料。速度通常比 selenium 快，但有些網站會特別拖慢 requests 請求的回應的速度。
+ 如果網站有提供 API，可直接用 API 抓資料。
+ 如果網站本身從資料庫抓資料用 API，可直接去找到 API 網址，然後用 requests 抓資料。

### selenium
+ 本意是測試網頁的模組，被用來當作爬蟲
+ 模擬人類的行為，比 requests 更像人，但並不代表不會被抓到
+ 通常只要想抓的資料用 JavaScript 包起來，而且在 Network 找不到那包資料，就只能用 selenium（因為沒有實際開一個瀏覽器，缺的資料不會補）
+ 就算 selenium 很厲害，也不能肆無忌憚，凡是要登入會員的網站，你亂來就還是被 ban


模組的使用時機，以遊戲來比喻的話

pandas 像是屬性相剋技，剛好有剋到效果就很好，沒剋到就沒效。

requests 像是速攻技，出招速度快，但如果遇到比較硬的對手（防爬機制強），就容易沒效果。

selenium 像是大招，出招較慢，幾乎能破防（幾乎任何防爬機制都能突破或繞過）。

## pandas 爬蟲：read_html
看到網頁裡面有表格，就要想到可以用 read_html。read_html 同時有送請求和解析的功能，送請求的部分只能用 get 請求，似乎沒什麼偽裝，所以可以直接放網址的網頁越來越少。但只要先用 requests 取得網頁原始碼，再用 read_html 解析，就可以省去用 BeautifulSoup 解析和定位的功夫。
回到剛才台鐵時刻表的例子，原本的程式碼如下：

````python
url = 'https://www.railway.gov.tw/tra-tip-web/tip/tip001/tip112/querybytime'
data = {'startStation': '1000-臺北', 'endStation': '3300-臺中', 'transfer': 'ONE', 'rideDate': '2025/05/28',
        'startTime': '00:00', 'endTime': '23:59', 'trainTypeList': 'ALL'}
r = requests.post(url, headers=header, data=data)
soup = BeautifulSoup(r.text, 'html.parser')

data = {}
for row in soup.select('tbody tr[class="trip-column"]'):
    train_number = row.select_one('td li').text.replace('\n', '')
    start_time = row.select_one('td:nth-child(2)').text
    end_time = row.select_one('td:nth-child(3)').text
    duration = row.select_one('td:nth-child(4)').text
    data[train_number] = [train_number, start_time, end_time, duration]

df = pd.DataFrame(data).transpose()
df
````

````python
# 還需要定位出自己想要的資料在哪，但如果改成：
url = 'https://www.railway.gov.tw/tra-tip-web/tip/tip001/tip112/querybytime'
data = {'startStation': '1000-臺北', 'endStation': '3300-臺中', 'transfer': 'ONE', 'rideDate': '2025/05/28',
        'startTime': '00:00', 'endTime': '23:59', 'trainTypeList': 'ALL'}
r = requests.post(url, headers=header, data=data)
df = pd.read_html(r.text)[0] # read_html會把所有表格都抓出來，想要的資料在第一個表格
df = df[df['出發時間'].str.match('\d{2}:\d{2}(?!.)')] # 由於台鐵時刻表的表格特別髒亂，需要做一些篩選。不懂沒關係，示範用而已
df = df[['車種車次 (始發站 → 終點站)', '出發時間', '抵達時間', '行駛時間']]
df

# 整個清單就直接整理好了
````

## 下載 driver

以 chromedriver 為例，chromedriver 的版本需要和瀏覽器版本一致（主版本號一致就好，例如目前是136）才可以使用。但如果你剛剛是用 pip 安裝 selenium，你的版本號會是 4.32，你可以直接跳過這個步驟，如果你在以前已經安裝過比較舊的 selenium 版本，建議更新。如果你在以前已經抓過比較舊的 driver，建議刪掉。

在 4.6 之後，selenium 會自動偵測有沒有 driver，如果沒有的話就會自動下載到

(Windows) C:\Users\使用者\.cache\selenium\chromedriver\win64

(Mac) $HOME\.cache\selenium

(Linux) $HOME\.cache\selenium

因此使用者再也不用管 driver 版本跟瀏覽器版本是否有一致。缺點是他不會刪掉舊的 driver，所以有時候還是要來刪一下舊的 driver。
